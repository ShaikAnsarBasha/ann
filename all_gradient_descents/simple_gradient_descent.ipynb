```python
import numpy as np
X = [[1],
     [2],
     [3],
     [4],
     [5]]
y = [5, 7, 9, 11, 13]
X = np.array(X)
y = np.array(y)
```

```python
def cost_mse(y, y_pred):
    n = y.shape[0]
    cost = (1/n)*sum((y-y_pred)**2)
    return cost    
```

```python
def gradient_descent_mse(X, y, epochs):
    weights = np.random.rand(1, 1)-0.5
    biases = np.random.rand(1,)-0.5
    learning_rate = 0.05
    n = X.shape[0]  
    
    for i in range(epochs):
        output_z = X.dot(weights.T) + biases
        
        cost = cost_mse(y, output_z)
        
        # here y, y_hat are interchanged viceversa to logisticregression
        output_errors = y.reshape(-1, 1) - output_z
        
        gradients = (-2/n)*(output_errors.T.dot(X))
        gradients_biases = (-2/n)*sum(output_errors)
        
        weights = weights - learning_rate*gradients
        biases = biases - learning_rate*gradients_biases
        
        print(f'weigths: {weights} biases: {biases} cost: {cost}')
        
    return weights, biases        
```

```python
gradient_descent_mse(X, y, epochs=1000)
```

```python
def sigmoid(z):
    z = 1 / (1 + np.exp(-z))
    return z
```

```python
def cost_logloss(y, y_pred):  # logloss
    epsilon = 1e-15
    n = y.shape[0]
    
    y_pred = np.where(y_pred==1, 1-epsilon, y_pred)
    y_pred = np.where(y_pred==0, epsilon, y_pred)

    cost = (-1/n)*sum(((y*np.log(y_pred)) + (1-y)*(np.log(1-y_pred))))
    return cost
```

```python
# we use logloss or binarycrossentropy for logistic regression or binary classification problems
def gradient_descent_logloss(X, y, epochs):
    
    weights = np.random.rand(1, 2)-0.5
    biases = np.random.rand(1,)-0.5
    learning_rate = 0.05
    n = X.shape[0]
    
    for i in range(epochs):
        output_z = X.dot(weights.T) + biases
        output_a = sigmoid(output_z)
        
        cost = cost_logloss(y.reshape(-1, 1), output_a)
        if cost <= 0.4733:
            break
        
        output_errors = output_a - y.reshape(-1, 1)
        
        gradients = (1/n)*(output_errors.T.dot(X))
        gradients_biases = (1/n)*np.sum(output_errors)

        weights = weights - learning_rate*gradients
        biases = biases - learning_rate*gradients_biases

        print(f'Epoch:{i}, weights: {weights}, biases: {biases}, cost: {cost}')

    return weights, biases
```

```python
gradient_descent_logloss(X_train ,y_train, 5000)
```

```python
model = keras.Sequential([
    keras.layers.Dense(1, input_shape=(2,), activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
model.fit(X_train, y_train, epochs=5000)
```

```python
# from above we got approximately same weights
```

```python
model.evaluate(X_test, y_test)
```
