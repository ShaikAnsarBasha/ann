```python
X = df.drop(60, axis=1)
y = df[60]
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
X_train.head()
```

# Using Deep Learning Model

## Model without Dropout Layer

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train.values, y_train.values, epochs=100, batch_size=8)
```

```python
model.evaluate(X_test.values, y_test.values)
```

```python
y_pred = model.predict(X_test.values).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test, y_pred))
```

### Model with Dropout Layer

```python
# if we use input_dim we use number '60'
# if we use input_shape we use matrix notation '(60,)'
modeld = keras.Sequential([
    
    # adding 60 hidden layers 
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    
    # after adding 60 hidden neurons 50%(30neurons) dropped randomly in every epoch
    keras.layers.Dropout(0.5),
    
    # now added 30neurons 
    keras.layers.Dense(30, activation='relu'),
    
    # 20% of neurons are dropped in 30neurons randomly 
    keras.layers.Dropout(0.2),
    
    # 15 hidden neurons are added
    keras.layers.Dense(15, activation='relu'),
    
    # out of 15 neurons 10% neurons are randomly
    keras.layers.Dropout(0.1),
    
    # output layer contains 1 neuron with sigmoid activation
    keras.layers.Dense(1, activation='sigmoid')
])
modeld.summary()
```

```python
modeld.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

modeld.fit(X_train.values, y_train.values, epochs=100, batch_size=8)
```

```python
modeld.evaluate(X_test.values, y_test.values)
```

Training Accuracy is still good but Test Accuracy Improved

```python
y_pred = modeld.predict(X_test.values).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test, y_pred))
```

```python
# from this we can conclude that by adding regulirazition we can increase accuracy of test data set
# it increased from 0.71 to 0.77
```
